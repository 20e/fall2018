```python
@ipywidgets.interact(gridsize = [4, 8, 16, 32, 64, 128],
                     xfrac = (0.05, 0.5, 0.05),
                     yfrac = (0.05, 0.5, 0.05))
def make_fun_hist(gridsize = 16, xfrac = 0.25, yfrac = 0.25):
    
    ax1 = plt.axes( [0.0, 0.0, 1.0 - yfrac, 1.0 - xfrac])
    ax2 = plt.axes( [1.0 - yfrac, 0.0, yfrac, 1.0 - xfrac], sharey = ax1)
    ax3 = plt.axes( [0.0, 1.0 - xfrac, 1.0 - yfrac, xfrac], sharex = ax1)

    ax1.hexbin(x, y, gridsize = gridsize)
    ax2.hist(y, orientation='horizontal', bins = gridsize)
    ax2.set_frame_on(False)
    ax3.hist(x, orientation='vertical', bins = gridsize)
    ax3.set_frame_on(False)
```
